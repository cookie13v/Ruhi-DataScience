```python
import pandas as pd

df = pd.read_csv('/content/EDA-DP.csv')
display(df.head())
```

I imported the CSV file and then created a Dataframe.

```python
df.info()
```

Looking at the datatypes i can now interpret and analyse the Data better.

```python
print('Missing values per column:')
display(df.isnull().sum())
```

Since there are no missing values, i can now start analysing.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(y=df['salary'])
plt.title('Box Plot of Salary')
plt.ylabel('Salary')
plt.show()
```

The boxplot show outliers, especially on the extreme top whisker. Therefore, this outlier needs to be handled as it can create problems in analysing in the future.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['salary'], bins=50, kde=True)
plt.title('Histogram of Salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

The histogram shows that the vast majority of salaries are concentrated at the lower end of the range, with very high salaries being extremely rare. The distribution is highly right-skewed, meaning most employees earn relatively modest salaries while a few earn extremely large amounts. This skewness suggests the presence of substantial outliers and indicates that the mean salary may be misleading without transformation or outlier handling.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 7))
sns.boxplot(x='experience_level', y='salary', data=df, palette='viridis')
plt.title('Salary Distribution by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
